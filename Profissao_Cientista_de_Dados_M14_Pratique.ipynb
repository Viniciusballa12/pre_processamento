# 1 - Carregue a base, verifique se os tipos de dados estão de acordo com cada coluna e caso não estejam realize a transformação dos tipos de dados.

```python
#seu código aqui
df['Genero'] = df['Genero'].fillna('Unknown')

# Converter colunas categóricas
categoricas = ['Genero', 'Casado', 'Dependents', 'Servico_Internet', 
               'Servico_Seguranca', 'Suporte_Tecnico', 'StreamingTV', 'Tipo_Contrato', 
               'PaymentMethod', 'Churn']
for coluna in categoricas:
    df[coluna] = df[coluna].astype('category')

# Converter 'Idoso' para int64 (substituir NaN por 0 se houver)
df['Idoso'] = df['Idoso'].fillna(0).astype('int64')

# Converter 'Tempo_como_Cliente' para int64 (substituir NaN por 0 se houver)
df['Tempo_como_Cliente'] = df['Tempo_como_Cliente'].fillna(0).astype('int64')

# Converter 'Pagamento_Mensal' e 'Total_Pago' para float64, tratando valores ausentes
df['Pagamento_Mensal'] = pd.to_numeric(df['Pagamento_Mensal'], errors='coerce')
df['Total_Pago'] = pd.to_numeric(df['Total_Pago'], errors='coerce')

# Verificar os novos tipos de dados
print("Novos tipos de dados após transformação:")
df.dtypes
print("\n")

# Exibir as primeiras linhas para confirmar
print("Primeiras linhas após transformação:")
df.head()
```

```python
df.dtypes
```

Explicação das Transformações
Colunas Categóricas: Usamos category para reduzir o uso de memória e facilitar análises estatísticas (ex.: contagens, proporções).

Colunas Numéricas:
Idoso e Tempo_como_Cliente foram mantidos ou convertidos para int64, pois representam contagens discretas.
Pagamento_Mensal e Total_Pago foram convertidos para float64 para lidar com valores decimais (ex.: 29.85) e valores ausentes (convertidos para NaN).

Valores Ausentes:
Genero foi preenchido com "Unknown" para manter consistência.
Pagamento_Mensal e Total_Pago usam errors='coerce' para converter strings inválidas (como vazios) para NaN.

Por algum motivo o PhoneService não tinha virado float, então transformei ele em float aqui mesmo

```python
#seu código aqui
# Tratar strings vazias como NaN e substituir conforme etapa anterior
df = df.replace(r'^\s*$', pd.NA, regex=True)
df['PhoneService'] = df['PhoneService'].fillna(df['PhoneService'].mode()[0])
df['Pagamento_Mensal'] = pd.to_numeric(df['Pagamento_Mensal'], errors='coerce').fillna(df['Pagamento_Mensal'].median())
df['Total_Pago'] = pd.to_numeric(df['Total_Pago'], errors='coerce').fillna(df['Total_Pago'].median())

# Converter 'PhoneService' para float
df['PhoneService'] = pd.to_numeric(df['PhoneService'].replace({'Yes': 1.0, 'No': 0.0}), errors='coerce').astype('float')

# Ajustar a lista de colunas categóricas com base nas colunas reais
colunas_categoricas = [col for col in ['Genero', 'Casado', 'Dependents', 'PhoneService', 'Servico_Internet', 
                                      'Servico_Seguranca', 'Suporte_Tecnico', 'StreamingTV', 'Tipo_Contrato', 
                                      'PaymentMethod', 'Churn'] if col in df.columns]


# Função para padronizar e corrigir
def padronizar_coluna(coluna):
    if coluna in df.columns:
        # Aplicar .str.lower() apenas se a coluna for de texto (object ou category)
        if df[coluna].dtype in ['object', 'category']:
            df[coluna] = df[coluna].str.lower()
        
        # Correções específicas por coluna
        elif coluna == 'PhoneService':
            df[coluna] = df[coluna].replace({'yes': 1.0, 'no': 0.0}).astype('float')
        elif coluna in ['Casado', 'Dependents', 'Churn']:
            df[coluna] = df[coluna].replace({'yes': 'sim', 'no': 'nao'})
        elif coluna == 'Servico_Internet':
            df[coluna] = df[coluna].replace({'no internet service': 'no_internet_service'})
        elif coluna in ['Servico_Seguranca', 'Suporte_Tecnico', 'StreamingTV']:
            df[coluna] = df[coluna].replace({'no internet service': 'no_internet_service', 'yes': 'sim', 'no': 'nao'})
        elif coluna == 'Tipo_Contrato':
            df[coluna] = df[coluna].replace({
                'month-to-month': 'month_to_month',
                'one year': 'one_year',
                'two year': 'two_year'
            })
        elif coluna == 'PaymentMethod':
            df[coluna] = df[coluna].replace({
                'electronic check': 'electronic_check',
                'mailed check': 'mailed_check',
                'bank transfer': 'bank_transfer',
                'credit card': 'credit_card'
            })

# Aplicar padronização
for coluna in colunas_categoricas:
    padronizar_coluna(coluna)

# Exibir as primeiras linhas para confirmar
print("Primeiras linhas após padronização:")
df[colunas_categoricas].head()
```
